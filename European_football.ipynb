##  Get each team match result


```python
def match_result(home_goals, away_goals):
    if (home_goals > away_goals):
        return 'Home'
    elif (home_goals < away_goals):
        return 'Away'
    else:
        return 'Tie'
matches_team = pd.read_sql_query('Select  T.team_long_name , m.home_team_goal, m.away_team_goal FROM Team as T , Match as M WHERE  T.team_api_id = M.home_team_api_id  ', con)

matches_team['result'] = matches_team.apply(lambda row: match_result(row['home_team_goal'],row['away_team_goal']), axis=1 )

matches_team 
```

## Get numbers of win, lost and draw matches of each team

- If a team has home goal > away goal , team is “win” in this match
- If a team has home goal < away goal , team is “lost” in this match
- If a team has home goal = away goal , team is “draw” in this match
- Use Case When statement

## After that get top 10 team with highest numbers of win matches

```python
#Câu hỏi 14: Liệt kê số trận thắng, thua và hòa của mỗi đội.
# Câu hỏi 15: Liệt kê 10 đội có số trận thắng nhiều nhất
win_lost_team =pd.read_sql(""" SELECT team_long_name,
sum(win) as win,
sum(lost) as lost,
sum(draw) as draw
FROM
( SELECT
Team.team_long_name as team_long_name,
(case when Match.home_team_goal > Match.away_team_goal then 1
else 0 end) as win,
(case when Match.home_team_goal < Match.away_team_goal then 1
else 0 end) as lost,
(case when Match.home_team_goal = Match.away_team_goal then 1
else 0 end) as draw
From Match INNER JOIN Team ON Match.home_team_api_id = Team.team_api_id
UNION ALL
SELECT 
Team.team_long_name as team_long_name,
(case when Match.away_team_goal > Match.home_team_goal then 1
else 0 end) as win,
(case when Match.away_team_goal < Match.home_team_goal then 1
else 0 end) as lost,
(case when Match.away_team_goal = Match.home_team_goal then 1
else 0 end) as draw
From Match INNER JOIN Team ON Match.away_team_api_id = Team.team_api_id

) as A
GROUP BY team_long_name
ORDER BY sum(win) desc

""", con)
win_lost_team.head(10)
```

##  Get percentage of each league to join the matches

```python
result = pd.read_sql("""
SELECT League.id as 'League ID', League.name as 'League name',
COUNT(league_id) AS 'Num_matches_per_league',
COUNT(Match.league_id)* 100.0 / (SELECT COUNT(Match.match_api_id) FROM Match) AS 'Percentage_match'
FROM Match
JOIN League ON League.id = Match.league_id
GROUP BY League.id;
""", con)
result
```

## Get percentage of score in each league

```python
result = pd.read_sql("""
SELECT League.name as 'League name',
SUM(home_team_goal + away_team_goal)* 100.0 / (SELECT SUM(home_team_goal + away_team_goal) FROM Match) AS 'Percentage_goals'
FROM Match
JOIN League ON League.id = Match.league_id
GROUP BY League.id;
""", con)
result
```

##  Get players who had overall_rating larger than 80

```python
player_rate =  pd.read_sql_query(
"""
SELECT P.Player_name ,MAX(PA.overall_rating)
FROM Player as P, Player_Attributes as PA 
Where P.player_api_id = PA.player_api_id AND PA.overall_rating > 80
GROUP BY P.Player_name

"""
, con)
player_rate
```
